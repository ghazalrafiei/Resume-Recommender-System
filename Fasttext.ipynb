```python
def normaz(t):
  _normalizer = Normalizer()
  return _normalizer.normalize(t)
```

```python
def tokenizePersianText(text):
  x = normaz(text)
  x = x.replace('\u200c',' ')
  x = x.replace('\u202a',' ')
  x = x.replace('\u202c',' ')
  x = x.replace('\u202b',' ')
  x = x.replace('\u202c',' ')

  # ^((\\u202a)|(\\u202c)|(\\x0c)|(\\uf0b7)|\\ue812|\\u200c)
  tz = Tokenizer()
  tokens = tz.tokenize_words(x)
  # s = SpellCheck()
  # tokens = [s.spell_corrector(w) for w in text]

  tokens = [w for w in tokens if not w in stop_words if len(w)>3 and w!='C']
  for t in tokens:
    if re.match(r'[a-zA-Z]',t):
      if t.lower() in eng_to_fa.keys():
        tokens.remove(t)
        tokens.append(eng_to_fa[t.lower()])
  return tokens
```

```python
import numpy as np
desc_model = np.mean([np.array(model.get_word_vector(w)) for w in desc_tokens],axis = 0)
```

```python
resumes_model = {}
for i in range(1,35):
  resumes_model[i] = np.mean([model.get_word_vector(w) for w in resumes[i]],axis = 0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial import distance

similarities = {}
sim_ec = {}
for i in range(1,35):
  similarities[i] = cosine_similarity(np.array(resumes_model[i]).reshape(1,-1),np.array(desc_model).reshape(1,-1))
  sim_ec[i] = distance.euclidean(np.array(resumes_model[i]).reshape(1,-1),np.array(desc_model).reshape(1,-1))
```

```python
ranking = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
result = pd.DataFrame()
result['number'] = [int(r[0]) for r in ranking]
result['score'] = [r[1].flatten()[0] for r in ranking]
result
```
